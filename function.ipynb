```python
def cal():
    l = int(input("Enter l = "))       #local variable
    b = int(input("Enter b = "))       #local varable
    h = int(input("Enter h = "))       #local variable
    global a,v
    a = l*b
    v = a*h
    print(a)
    print(v)
    
cal()
print(a,v)
```

```python
def cal(l,b,h):
    a = l*b
    v = a*h
    print(a)
    print(v)
    
l = int(input("Enter l = "))       #local variable
b = int(input("Enter b = "))       #local varable
h = int(input("Enter h = "))       #local variable
cal(l,b,h)
```

```python
def cal(y):
    l,b,h = y
    a = l*b
    v = a*h
    print(a)
    print(v)
    
l = int(input("Enter l = "))       #local variable
b = int(input("Enter b = "))       #local varable
h = int(input("Enter h = "))       #local variable
x = [l,b,h]
cal(x)
```
